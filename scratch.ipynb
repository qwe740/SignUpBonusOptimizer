```python
import numpy as np
from bonusOptimization import BonusOptimizer, HelperFunctions
```

```python
# Ideal bonus match for odds at a given vig
vig = 0.03 #Standard vig for -110, -110
bonus_odds = np.array([-500,-400,-300,-200,100,200,300,400,500,600,700,800,900,1000])
f1 = np.vectorize(HelperFunctions.otherOddsGivenVig)
hedgeodds_est = f1(bonus_odds,vig)
rfbet_ev = [BonusOptimizer.RFBetEV(1000,bonus_odds[i],hedgeodds_est[i]) for i in range(len(bonus_odds))]
bonusmatch_ev = [BonusOptimizer.BonusMatchEV(1000,bonus_odds[i],hedgeodds_est[i]) for i in range(len(bonus_odds))]
```

```python
import matplotlib.pyplot as plt
plt.plot(bonus_odds,rfbet_ev,label='Risk Free Bet')
plt.plot(bonus_odds,bonusmatch_ev,label='Bonus Match')
plt.legend()
plt.show()
```
